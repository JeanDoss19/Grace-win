```python
import pandas as pd
import numpy as np
import seaborn; seaborn.set()
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
pd.read_csv("Fish.csv")
```

```python
df= pd.read_csv("Fish.csv")
df.head()
```

```python
df.drop('Length1', axis=1)
```

```python
sns.lmplot(x='Weight', y='Height', data=df)
```

```python
sns.lmplot(x='Weight', y='Height', data=df, 
           fit_reg=False, 
           hue='Width')   
```

```python
plt.subplots(figsize=(24,18))
sns.violinplot(x='Species', y='Length3', data=df)
```

```python
sns.boxplot(data=df)
```

```python
plt.figure(figsize=(22,18))

#swarm plot of Stats from melted dataframe
sns.swarmplot(x='Species', y='Length2', 
              data=df,
              hue='Weight')
```

```python
#show correlation of stats via heatmap
corr = df.corr()

sns.heatmap(corr, vmin=-1, annot=True)
```
